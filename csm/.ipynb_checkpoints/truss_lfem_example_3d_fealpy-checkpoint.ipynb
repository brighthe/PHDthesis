```python
from fealpy.pde.truss_model import Truss_3d
from fealpy.functionspace import LagrangeFESpace
from fealpy.fem import BilinearForm
from fealpy.fem import TrussStructureIntegrator
from fealpy.fem import DirichletBC
from scipy.sparse.linalg import spsolve

import matplotlib.pyplot as plt
import numpy as np
```

```python
pde = Truss_3d()
mesh = pde.init_mesh()
NN = mesh.number_of_nodes()
NC = mesh.number_of_cells()
print("NN:", NN)
print("NC:", NC)

fig = plt.figure()
axes = fig.add_subplot(111, projection='3d') 
mesh.add_plot(axes)
plt.show()
```

```python
GD = mesh.geo_dimension()
print("GD:", GD)
space = LagrangeFESpace(mesh, p=1, doforder='vdims')
gdof = space.number_of_global_dofs()
ldof = space.number_of_local_dofs()
print("gdof:", gdof)
print("ldof:", ldof)
uh = space.function(dim=GD) 
print("uh:(gdof, GD)\n", uh)
print(uh.shape)
```

```python
vspace = GD*(space, ) # 把标量空间张成向量空间
print("space:", vspace)
bform = BilinearForm(vspace)
print("bform:", bform)
```

```python
E = pde.E # 杨氏模量
A = pde.A # 横截面积
print("单元刚度矩阵 k:", 
      TrussStructureIntegrator(E, A).assembly_cell_matrix(vspace).shape, 
      "(NC, GD*ldof, GD*ldof)")
bform.add_domain_integrator(TrussStructureIntegrator(E, A))
print("K1", bform.add_domain_integrator(TrussStructureIntegrator(E, A)))
K = bform.assembly() # 基函数由标量函数组合而成的向量函数空间
print("K2:", bform.assembly())
```

```python
 # 加载力的条件 
F = np.zeros((uh.shape[0], GD), dtype=np.float64)
idx, f = mesh.meshdata['force_bc'] # idx.shape = (2, ), f.shape = (3, )
print("F:(gdof GD)\n", F)
print("idx:", idx)
print("f\n:", f)
print("F[idx]\n:", F[idx])
F[idx] = f # (10, 3)
print("F[idx]\n:", F[idx])
print("F\n:", F)
```

```python
idx, disp = mesh.meshdata['disp_bc']
print("idx:", idx)
print("disp:", disp)
bc = DirichletBC(vspace, disp, threshold=idx)
A, F = bc.apply(K, F.flat, uh)
print("A:(GD*gdof, GD*gdof)\n", A.toarray())
print("F:(GD*gdof)\n", F)
```

```python
uh.flat[:] = spsolve(A, F)
print("uh:", uh)
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1, projection='3d') 
mesh.add_plot(axes)
scale = 100 # 变形系数
mesh.node += scale*uh
mesh.add_plot(axes, nodecolor='b', cellcolor='m')
plt.show()
```
