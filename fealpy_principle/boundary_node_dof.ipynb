边界上的节点

```python
from fealpy.mesh.triangle_mesh import TriangleMesh
mesh = TriangleMesh.from_box(box=[0, 1, 0, 1], nx=3, ny=3)
import matplotlib.pyplot as plt
fig = plt.figure()
axes = fig.gca()
mesh.add_plot(axes)
mesh.find_node(axes, showindex=True, 
               color='k', marker='o', markersize=8, fontsize=16, fontcolor='r')
plt.show()

isBdNode = mesh.ds.boundary_node_flag()
print("isBdNode:", isBdNode)
```

边界上的自由度

```python
import numpy as np
from fealpy.functionspace import LagrangeFESpace
p = 2

space = LagrangeFESpace(mesh, p=p, spacetype='C', doforder='vdims')
isBdDof = space.is_boundary_dof()
print("isBdDof:", isBdDof)
```

插值点绘图

```python
points = mesh.interpolation_points(p=p)
ipoints = space.interpolation_points() 

print("error:", np.sum(np.abs(points-ipoints)))

fig = plt.figure()
axes = fig.gca()
mesh.add_plot(axes)
mesh.find_node(axes, node=points, showindex=True, 
               color='k', marker='o', markersize=8, fontsize=16, fontcolor='r')
plt.show()
```

边界点绘图

```python
dflag = mesh.ds.boundary_node_flag()
print("dflag:", dflag)
node = mesh.entity('node')
bd_points = node[dflag]

import matplotlib.pyplot as plt
fig = plt.figure()
axes = fig.gca()
mesh.add_plot(axes)
mesh.find_node(axes, node=bd_points, showindex=True, 
               color='k', marker='o', markersize=8, fontsize=16, fontcolor='r')
plt.show()
```
