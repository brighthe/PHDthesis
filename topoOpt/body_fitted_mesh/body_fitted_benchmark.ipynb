```python
import matplotlib.pyplot as plt

from domain_2d import RectangleDomain, BoxWithCircleHolesDomain
from fealpy.mesh import TriangleMesh
```

```python
domain = RectangleDomain(domain=[0, 1, 0, 1], hmin=0.1, hmax=0.3)
maxiter = 500
mesh = TriangleMesh.from_domain_distmesh(domain, maxit=maxiter)
fig = plt.figure()
axes = fig.gca()
mesh.add_plot(axes)
plt.show()  
```

```python
circles= [(0.25, 0.5, 0.1), (0.75, 0.5, 0.1)]
domain = BoxWithCircleHolesDomain(box=[0, 1, 0, 1], \
                                circles=circles, hmin=0.03, hmax=0.1)
maxiter = 500
mesh = TriangleMesh.from_domain_distmesh(domain, maxit=maxiter)
fig = plt.figure()
axes = fig.gca()
mesh.add_plot(axes)
plt.show()  
```
